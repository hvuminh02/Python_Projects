EXPLORATORY DATA ANALYSIS IN PANDAS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'world_population.csv')
df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#Overall information of the dataset
df.info()
```

```python
#Describe overall the dataset
df.describe()
```

```python
#Explore how many values is null
df.isnull().sum()
```

```python
#Explore how many unique values
df.nunique()
```

```python
# Top 10 2022 Population
df.sort_values(by = "2022 Population", ascending = False).head(10)
```

```python
#Top10 World Population Percentage
df.sort_values(by = "World Population Percentage", ascending = False).head(10)
```

```python
#Find the pairwise correlation of all columns
df.corr()
```

```python
#Heatmap using seaborn
sns.heatmap(df.corr(), annot = True)
    #change the size of the heatmap
plt.rcParams['figure.figsize'] = (25, 7)
plt.show()
```

```python
#Grouping Data

df.groupby('Continent').mean()
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by = "2022 Population", ascending= False)
#the other way
#df2 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by = "2022 Population", ascending= False)
df2
```
